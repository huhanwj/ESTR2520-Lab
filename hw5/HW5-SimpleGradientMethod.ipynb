## Evaluating the objective function & the gradient vector

Program the objective function:
$$ f(x) = \log(1 + e^{x_1}) + \frac{1}{2}(x_1 - x_2)^2 + \frac{1}{4} x_1^4$$
and its gradient at any $x \in \mathbb{R}^2$ into Julia functions.

```julia
function obj_fct(x)
    # this function should take a 2-dim. vector "x" as the input, and return the objective value.
    func_v = log1p(exp(x[1]))+0.5*(x[1]-x[2])^2+0.25*x[1]^4
    return func_v
end
```

```julia
function grad_fct(x)
    # this function should take 2-dim. vector "x" as the input, and return the gradient (also a 2-dim. vector) 
    # of "f" at "x".
    grad_v = [(exp(x[1]))/(1+exp(x[1]))+x[1]-x[2]+x[1]^3; x[2]-x[1]]
    return grad_v
end
```

With the objective function and gradient function coded, you may now program the gradient method as follows:

```julia
# initialize the variables 
x_init = [1.; 1.]; store_obj = []; push!(store_obj,obj_fct(x_init))

# begin the iteration
x = copy(x_init)

for iter = 1 : 10000 # we simply let the GD method run for 10000 iterations 
                    # - you may increase this number to get a more accurate solution
    # your code here - you may use a diminishing step size as gamma = 1/iter here 
    x = x - 1/iter*grad_fct(x)
    # the following function simply stores the current objective value
    push!( store_obj, obj_fct( x ) );
end
```

```julia
plot( store_obj , labels = "f(x)", lw=2, title = "Objective Value" , size=(500,300) )
```
